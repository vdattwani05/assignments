The main goal of this assignment consists of creating a simplified version of Splitser (WieBetaaltWat). 

Imagine we have a group of roommates: Tom, Vanessa, Chiara, Anna, and Natt. To fix this, they can use Splitser. 

Splitser is an app that lets each individual in a group log purchases and share the costs with the rest of the group. While Splitser allows the expenses to be shared unequally, in our version, the object is to split the total cost _**equally**_ among all the roommates.

Therefore, based on a set of purchases made by the roommates for the house, each roommate will either end up with a credit, a debit, or a €0 final balance.

````python
#// BEGIN_TODO [get_individual_totals] Get individual totals 

def get_individual_totals(numbers:int, purchases:float=[]) -> list[float]:
    '''Inputs number of users in the household in numbers and creates an array (total) to store 
    total expenditure by each member on their respective index.
    
    2 valid inputs:
    
    >>> get_individual_totals(3, [5.00, 0, 29.75, 0])
    [34.75, 0.0, 0.0]
    
    >>> get_individual_totals(5, [35.00, 1, 13.45, 0, 99.85, 4, 67.8,3])
    [13.45, 35.0, 0.0, 67.8, 99.85]
    
    3 inputs that do not satisfy the requirements:
    
    >>> get_individual_totals(5, [35.00, 1.3, 13.45, 0, 99.85, 4, 67.8,3])
    Invalid input: Wrong concatenation of costs and ids in parameter 'purchases'
    []
    
    >>> get_individual_totals(5, [35.00, 1, 13.45, 0, 99.85, 7, 67.8,3])
    Invalid input: ID number invalid.
    []
    
    >>> get_individual_totals(5, [-35.04, 1, 13.45, 0, 993.85, 4, 67.8,3])
    Invalid input: Cannot have negative value expenditure.
    []
    
    '''
    
    
    #Declaring an array 
    totals: int = [0.0]*numbers
    
    try:
        
        for i in range(1, len(purchases), 2):
            #print(purchases[i])
            ID_no: float = purchases[i]
            ID_int:int = int(ID_no)
            if ID_no != ID_int:
                raise TypeError()
                
            if purchases[i] > numbers-1:
                raise IndexError()
            
            if numbers < 0:
                raise IndexError()
        
        for i in range(0, len(purchases),2):
            if purchases[i] < 0:
                raise ValueError()
            
        


        for num in range(numbers):

            for digit in range(0,len(purchases)):
                if purchases[digit] == num:
                    totals[num] += purchases[digit-1]

        return totals
    
    except TypeError:
        print("Invalid input: Wrong concatenation of costs and ids in parameter 'purchases'")
        return []
        
    except IndexError:
        print("Invalid input: ID number invalid.")
        return []
    
    except ValueError:
        print("Invalid input: Cannot have negative value expenditure.")
        return []
    
#// END_TODO [get_individual_totals]

get_individual_totals(2, [15.00, 0, 23.15, 1.20])
````

## Task 2: Final Balances

We then create a second function that 
- takes the total spent by each individual: a list of floats, where the float number at index `i` corresponds to the total spent by the person with ID `i`
- returns the final balance of each individual: a list of floats, where the float number at index `i` corresponds to the final balance of the person with ID `i`

**Notes**:
* The final balance of an individual can be calculated by doing `spent - avg`, where
    * `spent` is the total amount spent by the individual
    * `avg` is the total amount spent by the group, divided by the number of people in the group
* Remember to deal with the possibily that the input list is empty
* Do not forget to write the **type hints** and **docstring** of the function.
* Include at least _two_ (2) `doctest` examples

**Examples**:

*Input:*  
```python
get_individual_final_balances([10.0, 20.0, 3.0])
```

*Output:*  

```python
[-1.0, 9.0, -8.0]
```

````python
#// BEGIN_TODO [get_individual_final_balances] Get individual final balances

def get_individual_final_balances(individual_spent:float=[]) -> list[float]:
    '''Inputs number of users in the household in numbers and creates an array (total) to store 
    total expenditure by each member on their respective index.
    
    2 expected inputs:
    
    >>> get_individual_final_balances([10.0, 20.0, 3.0])
    [-1.0, 9.0, -8.0]
    
    >>> get_individual_final_balances([])
    Invalid input: List is empty.
    
    
    '''
    
        
    avg: float = 0
    spent: float = 0
    
    try:
        for individual in individual_spent:
            spent += individual
            
        if spent == 0:
            raise ZeroDivisionError()
            
        avg = spent/len(individual_spent)

        final_balance: float = [0.0]*len(individual_spent)

        for person in range(len(individual_spent)):
            final_balance[person]= individual_spent[person]-avg

        return final_balance
    
    except ZeroDivisionError:
        print('Invalid input: List is empty.')
        
#// END_TODO [get_individual_final_balances]

get_individual_final_balances([10.0, 20.0, 3.0])
````

### 3.2: Data Conversion
We now create another function that instead of two parameters only takes 1 parameter:

This means that the number of people and their purchases will be merged into one list of strings and it is our job to figure out how to
- transform the list into parameters that are fit for `simplified_splitser`
- **print** the output of `simplified_splitser` as multiple lines with 
    - each line having format "[name]: € [cost]" and 
    - the cost having **2 decimal points** (.2f).

**Notes:** 

- To check if a string `my_string` can be converted to an int, you can use `my_string.isdigit()`, this outputs a boolean. 
- To check if a string `my_string` can be converted to a float, you can use `my_string.replace('.', '', 1).isdigit()`, this outputs a boolean.

````python
#// BEGIN_TODO [string_list_simplified_splitser] Convert single string list into parameters for simplified_splitser

def string_list_simplified_splitser(string_list:str=[]) -> list[str]:
    '''
    The function takes a string input and performs the necessary checks to do so, it follows the operations of Task 3.1.
    
    >>> string_list_simplified_splitser(["5", "12.45", "1", "15.6", "4", "25.5", "2", "4.30", "1"])
    0: € -11.57
    1: € 5.18
    2: € 13.93
    3: € -11.57
    4: € 4.03

    >>> string_list_simplified_splitser(["3", "122.45", "1", "18.6", "2", "95.5", "2", "2.0", "0"])
    0: € -78.18
    1: € 42.27
    2: € 35.92
    
    >>> string_list_simplified_splitser(["2", "122.45", "2", "18.6", "2", "95.5", "2", "2.0", "0"])
    Invalid input: ID number invalid.
    Invalid input: List is empty.
    '''
    
    
    num_list: float = [0.0]*len(string_list)
    
    for string in range(len(string_list)):
        mystring = string_list[string]
        if string == 0:
            if mystring.isdigit():
                num_list[string] = int(string_list[string])
        else:
            if mystring.isdigit() or mystring.replace('.', '', 1).isdigit():
                num_list[string] = float(string_list[string]) 

    num_of_people: int = num_list[0]
    list_values = num_list[1:]
    #return list_values
    
    sim_convert = simplified_splitser(num_of_people, list_values)
    
    if sim_convert != None:
        for money in range(len(sim_convert)):
            print(f'{money}: € {"{:.2f}".format(sim_convert[money])}')
        
        
    
#// END_TODO [string_list_simplified_splitser]


string_list_simplified_splitser(
    ["5", "12.45", "1", "15.6", "4", "25.5", "2", "4.30", "1"]
)
````
